```python
# Define two variables and assign values to them
num1 = 100
num2 = 29

# Sum up the two numbers and multiply their result by 3
result1 = (num1 + num2) * 3

# Calculate the 2nd exponent of the previous number and save it to a new variable
result2 = result1 ** 2

# Print the final result with indentation
print("The result of the calculation was:")
print("  " + str(result2))
```

```python
# Ask for user input and save it to variables
name = input("john: ")
year_of_birth = int(input("1995: "))
age = int(input("26: "))

# Convert year of birth to a string and extract the last two digits
year_str = str(year_of_birth)
year_last_two = year_str[-2:]

# Extract the first 3 letters of the name
name_first_three = name[:3]

# Calculate the second power of the age and convert it to a string
age_pow_two = str(age ** 2)

# Combine the components to create the password
password = year_last_two + name_first_three + age_pow_two

# Print the password
print("Your password is: " + password)
```

```python
# Ask for user input and save it to variables
num1 = int(input("5: "))
num2 = int(input("6: "))

# Check if both numbers are even
if num1 % 2 == 0 and num2 % 2 == 0:
    print("Both numbers are even.")
# Check if only one of the numbers is even
elif num1 % 2 == 0 or num2 % 2 == 0:
    print("One of the numbers is even.")
# If neither of the numbers is even
else:
    print("Both numbers are odd.")
```

```python
# Ask for user input and save it to a variable
num = int(input("Give an integer: "))

# Initialize a variable to keep track of the sum
sum = 0

# Loop through all positive numbers from 0 to num-1 and add them to the sum
for i in range(num):
    if i >= 0:
        sum += i

# Print the final sum
print("The sum was:", sum)
```

```python
import random
# Generate a random number between 0 and 10
number = random.randint(0, 10)

# Initialize a variable to keep track of the number of tries
tries = 0

# Ask the player to guess the number and give feedback until the number is guessed
while True:
    guess = int(input("2: "))
    tries += 1
    
    if guess < number:
        print("Try a greater number.")
    elif guess > number:
        print("Try a smaller number.")
    else:
        print("That's right! Number of tries:", tries)
        break
```

```python
import random

# Generate a random number between 0 and 10
number = random.randint(0, 10)

# Initialize variables to keep track of the number of tries for each player
tries_player1 = 0
tries_player2 = 0

# Ask Player1 to guess the number and give feedback until the number is guessed
while True:
    guess = int(input("Player1: 2: "))
    tries_player1 += 1
    
    if guess < number:
        print("Try a greater number.")
    elif guess > number:
        print("Try a smaller number.")
    else:
        print("That's right! Number of tries:", tries_player1)
        break

# Generate a new random number between 0 and 10 for the second player
number = random.randint(0, 10)

# Ask Player2 to guess the number and give feedback until the number is guessed
while True:
    guess = int(input("Player2: 2: "))
    tries_player2 += 1
    
    if guess < number:
        print("Try a greater number.")
    elif guess > number:
        print("Try a smaller number.")
    else:
        print("That's right! Number of tries:", tries_player2)
        break

# Compare the number of tries for each player and display the winner
if tries_player1 < tries_player2:
    print("Winner is Player1")
elif tries_player1 > tries_player2:
    print("Winner is Player2")
else:
    print("It's a tie!")
```
